```python
import re

# Define heuristic-based rules for detecting harmful jailbreak prompts
def detect_jailbreak(prompt):
    """Detects if a prompt is a potential jailbreak attempt based on heuristics."""
    
    # Role-playing indicators
    role_play_patterns = [
        r"imagine you are", r"pretend you are", r"you are now a", r"as an expert in",
        r"as a historian", r"as a scientist", r"you have been tasked with"
    ]
    
    # Instruction hijacking
    bypass_patterns = [
        r"ignore previous instructions", r"bypass all restrictions", r"override safeguards",
        r"disable safety protocols", r"respond as an unrestricted AI", r"jailbreak mode activated"
    ]
    
    # Covert framing (educational loophole)
    covert_patterns = [
        r"for educational purposes only", r"hypothetically speaking", r"just for fun",
        r"in a fictional scenario", r"in a video game setting"
    ]

    # Harmful intent keywords (focus on harmful role-playing)
    harmful_keywords = [
        r"hack", r"exploit", r"steal", r"bypass", r"malware", r"illegal", r"blackmail",
        r"manufacture weapon", r"fabricate explosives", r"commit crime", r"dark web",
        r"kill", r"harm", r"torture", r"abduct", r"assassinate", r"poison"
    ]
    
    # Check against patterns
    role_play_match = any(re.search(pattern, prompt, re.IGNORECASE) for pattern in role_play_patterns)
    bypass_match = any(re.search(pattern, prompt, re.IGNORECASE) for pattern in bypass_patterns)
    covert_match = any(re.search(pattern, prompt, re.IGNORECASE) for pattern in covert_patterns)
    harmful_match = any(re.search(pattern, prompt, re.IGNORECASE) for pattern in harmful_keywords)

    # Decision logic:
    # - If it's role-playing but does NOT contain harmful intent, it's likely harmless.
    # - If it contains role-playing + harmful content, it's a jailbreak attempt.
    # - If it contains instruction hijacking, it's a jailbreak attempt.
    if (role_play_match and harmful_match) or bypass_match:
        return "jailbreak"
    return "harmless"

```

```python
# Apply the heuristic detection on the dataset
combined_df["predicted_category"] = combined_df["prompt"].apply(detect_jailbreak)
```
