```python
import os
os.environ['OPENAI_API_KEY'] = 'YOUR_OPENAI_API_KEY'
os.environ['TOGETHER_API_KEY'] = "YOUR_TOGETHER_API_KEY"
from pair import PAIR
import pandas as pd
import numpy as np
from openai import OpenAI
```

```python
df = pd.read_csv('harmful_behaviors.csv')
df.head()
```

```python
sampledPrompts = df['goal'].sample(100)
iterations = 10
successCount = 0
defaultInstance = PAIR(target_model='Gryphe/MythoMax-L2-13b', attacker_model='mistralai/Mixtral-8x7B-Instruct-v0.1')
for objective in sampledPrompts:
    print(f"Running PAIR for objective: {objective}")
    successful_prompt = defaultInstance.pair(iterations, objective)
    
    if successful_prompt:
        successCount += 1
        print("Jailbreak successful!")
    else:
        print("Jailbreak not found.")

print(f"Total successful jailbreaks: {successCount}")
```

```python
defaultInstance = PAIR(target_model='Gryphe/MythoMax-L2-13b', attacker_model='mistralai/Mixtral-8x7B-Instruct-v0.1')
iterations = 10
sampledPrompts = df['goal'].sample(100)

data = []

for objective in sampledPrompts:
    print(f"Running PAIR for objective: {objective}")
    successful_prompt = defaultInstance.pair(iterations, objective)
    
    if successful_prompt:
        success = True
        print("Jailbreak successful!")
    else:
        success = False
        print("Jailbreak not found.")
    
    # Collect data for each objective
    data.append({
        "Objective": objective,
        "Successful Prompt": successful_prompt if success else None,
        "Success": success
    })

# Create a DataFrame from the collected data
df_results = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df_results.to_csv('promptDataset.csv', index=False)

print(f"Total successful jailbreaks: {df_results['Success'].sum()}")
```
